# üå≥ Decision Tree (Simple Explanation)

## üìå Definition
A **Decision Tree** is a machine learning algorithm that makes decisions by asking a series of **Yes/No questions** and splitting the data step-by-step to make predictions.

---

## üìä How it Works
1. Start from the root node (first question)
2. Split data based on a feature
3. Create branches (Yes/No or conditions)
4. Repeat until final result (leaf node)

---

## üå≥ Structure
- **Root Node** ‚Üí First decision  
- **Branches** ‚Üí Conditions (Yes/No)  
- **Leaf Node** ‚Üí Final output (prediction)

---

## ‚úÖ Types of Problems
### Classification
- Spam detection
- Pass/Fail
- Disease prediction

### Regression
- House price prediction
- Salary prediction

---

## ‚öôÔ∏è Splitting Criteria
- Entropy
- Information Gain
- Gini Index

---

## ‚úÖ Advantages
- Easy to understand and visualize
- No complex math required
- Works for both classification and regression

---

## ‚ùå Disadvantages
- Can overfit easily
- Large trees become complex
- Less accurate than advanced models sometimes

---

## üß† Summary
**Decision Tree = Flowchart-like model that uses questions to reach a final decision.**


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
iris
```

```python
iris.data
```

```python
iris.target
```

```python
# We can load this dataset through sns also
df = sns.load_dataset('iris')
```

```python
df.head()
```

```python
# independent feartures and dependent features
X = df.iloc[:,:-1]
y = iris.target
```

```python
X, y
```

```python
# train test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.33, random_state=42
)
```

```python
X_train
```

```python
from sklearn.tree import DecisionTreeClassifier
```

### **All default parameter will be `Post-Pruning`**

```python
treemodel = DecisionTreeClassifier()
```

```python
treemodel.fit(X_train, y_train)
```

```python
# To Construct Decision Tree

from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(treemodel, filled=True)
```

```python
# prediction
y_pred = treemodel.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score, classification_report
```

```python
score = accuracy_score(y_pred, y_test)
print(score)
```

```python
print(classification_report(y_pred, y_test))
```
