## **Univariate Analysis**

```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('./data/Univariate-analysis-titanic-train.csv')
```

```python
df.head()
```

### 1. Categorical Data

a. Countplot

```python
# sns.countplot(x=df['Survived'])
# df['Survived'].value_counts().plot(kind='bar')
sns.countplot(x=df['Pclass'])
```

b. Pie Chart

```python
df['Survived'].value_counts().plot(kind='pie', autopct="%.2f")
```

### 2. Numerical Data

a. Histogram

```python
import matplotlib.pyplot as plt
plt.hist(df['Age'])
plt.show()
```

b. Distplot

```python
sns.histplot(df['Age'], kde=True)
```

c. Boxplot

```python
sns.boxplot(df['Age'])
```
