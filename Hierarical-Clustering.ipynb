## **Hierarcial Clustering Implementation**

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
from sklearn import datasets
```

```python
# Import IRIS dataset
iris = datasets.load_iris()
```

```python
iris_data = pd.DataFrame(iris.data)
```

```python
iris_data.columns = iris.feature_names
```

```python
iris_data
```

```python
# Standardization
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_scaled = scaler.fit_transform(iris_data)
```

```python
X_scaled
```

```python
X_scaled.shape
```

```python
## Apply the PCA
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
```

```python
pca
```

```python
pca_scaled = pca.fit_transform(X_scaled)
```

```python
pca_scaled
```

```python
pca_scaled.shape
```

```python
plt.scatter(pca_scaled[:,0], pca_scaled[:,1])
```

### **Agglomerative Clustering**

```python
# To construct a dendogram
import scipy.cluster.hierarchy as sc

# plot the dendogram
plt.figure(figsize=(20, 7))
plt.title("Dendograms")

# create dendogram
sc.dendrogram(sc.linkage(pca_scaled, method='ward'))
plt.title("Dendogram")
plt.xlabel("Sample Index")
plt.ylabel("Eculedian Distance")
```

```python
from sklearn.cluster import AgglomerativeClustering
cluster = AgglomerativeClustering(n_clusters=2, metric='euclidean', linkage='ward')
cluster.fit(pca_scaled)
```

```python
cluster.labels_
```

```python
plt.scatter(pca_scaled[:,0], pca_scaled[:,1], c=cluster.labels_)
```

```python
# silhoutte score
from sklearn.metrics import silhouette_score
```

```python
silhouette_cofficients = []

for k in range(2,11):
    agglo = AgglomerativeClustering(n_clusters=k, metric="euclidean", linkage='ward')
    agglo.fit(X_scaled)
    score = silhouette_score(X_scaled, agglo.labels_)
    silhouette_cofficients.append(score)
```

```python
plt.plot(range(2,11), silhouette_cofficients)
plt.xticks(range(2,11))
plt.xlabel("Number of Clusters")
plt.ylabel("Silhoutte Coeffecient")
plt.show()
```
